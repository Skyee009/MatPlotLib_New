# 1. The Basics

# Easy

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y)
plt.show()
```

```python
# Add title testchart
plt.title("Test Chart")
plt.plot(x,y)
```

```python
# Show circular markers on the lines
plt.plot(x,y, marker = "o")
```

# Medium

```python
# Plot 2 lines on same graph
z=[9,8,7]
m=[6,5,4] 
plt.plot(x,y, marker = "^")
plt.plot(z,m, marker = "^")
```

```python
# Add labels on both axis
plt.xlabel("X label")
plt.ylabel("Y label")
z=[9,8,7]
m=[6,5,4] 
plt.plot(x,y, marker = "^")
plt.plot(z,m, marker = "^")
```

```python
# Display a legend
z=[9,8,7]
m=[6,5,4] 
plt.plot(x,y, marker = "^", label = "Sales")
plt.plot(z,m, marker = "^", label = "Profit")
plt.legend()
```
